```python
dframe['Glucose']=dframe['Glucose'].replace(0,dframe['Glucose'].median())
```

```python
dframe['BloodPressure']=dframe['BloodPressure'].replace(0,dframe['BloodPressure'].median())
```

```python
dframe['SkinThickness']=dframe['SkinThickness'].replace(0,dframe['SkinThickness'].median())
```

```python
x=dframe.iloc[:,:-1]
y=dframe.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
```

```python
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

# Logistic Regression

```python
LR=LogisticRegression()
model1=LR.fit(x_train_scaled,y_train)
```

```python
y_pred_train1=model1.predict(x_train_scaled)
y_pred_test1=model1.predict(x_test_scaled)
```

```python
print("Training Confusion Matrix\n",confusion_matrix(y_train,y_pred_train1))
print()
print()
print("Testing Confusion Matrix\n",confusion_matrix(y_test,y_pred_test1))
```

```python
print("Training Classification Report\n",classification_report(y_train,y_pred_train1))
print()
print()
print("Testing Classification Report\n",classification_report(y_test,y_pred_test1))
```

```python
print("Training Accuracy Score\n",accuracy_score(y_train,y_pred_train1))
print()
print()
print("Testing Accuracy Score\n",accuracy_score(y_test,y_pred_test1))
```
